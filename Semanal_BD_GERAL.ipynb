### MONTANDO OS DATAFRAMES PARA CADA REFERENCIA

```python
## FECHADO
q_fechado = f"""
DECLARE @data_ref DATE;
SET @data_ref = '{data_ref}'
SELECT * FROM selloutmensal_trade WHERE DATA = DATEADD(month,-1,@data_ref) """

result_fechado = conn.execute(q_fechado).fetchall()
# conn.description traz a lista do schema de cada coluna
columns = [column[0] for column in conn.description]

df_fechado = pd.DataFrame.from_records(result_fechado, columns= columns)

df_fechado['STATUS'] = 'FECHADO'
df_fechado.head()
```

```python
## MES ATUAL ANO ANTERIOR

q_mes_atual_ano_anterior = f"""
DECLARE @data_ref DATE;
SET @data_ref = '{data_ref}'
SELECT * FROM selloutmensal_trade
WHERE DATA = DATEADD(year,-1,@data_ref) """

result_mes_atual_ano_anterior = conn.execute(q_mes_atual_ano_anterior).fetchall()
df_mes_atual_ano_anterior = pd.DataFrame.from_records(result_mes_atual_ano_anterior, columns= columns)
df_mes_atual_ano_anterior['STATUS'] = 'MÊS ATUAL ANO ANTERIOR'
df_mes_atual_ano_anterior.head()
```

```python
## MES/ANO ANTERIOR
# dataa = datetime.strptime('2022/10/01','%Y/%m/%d').date()

q_mes_ano_anterior = f"""
DECLARE @data_ref DATE;
SET @data_ref = '{data_ref}'
SELECT * FROM selloutmensal_trade
WHERE DATA = DATEADD(month,-1,DATEADD(year,-1,@data_ref)) """

df_mes_ano_anterior = pd.read_sql(q_mes_ano_anterior, conexao)

df_mes_ano_anterior['STATUS'] = 'MÊS/ANO  ANTERIOR'
df_mes_ano_anterior.head()
```

```python
## PROJEÇÃO M1

q_projecao_m1 = f"""
DECLARE @data_ref DATE;
SET @data_ref = '{data_ref}'
SELECT * FROM selloutmensal_trade
WHERE DATA = DATEADD(month,1,@data_ref)"""

df_projecao_m1 = pd.read_sql(q_projecao_m1, conexao)

df_projecao_m1['STATUS'] = 'PROJEÇÃO M1'
df_projecao_m1.head()
```

#### JUNTANDO OS DATA FRAMES

```python
df1 = pd.concat([df_fechado,df_mes_atual_ano_anterior,df_mes_ano_anterior,df_projecao_m1])
len(df1)
#df1.head()
```

```python
# UPPER EM TODAS AS COLUNAS
# df1.dtypes
df1['BANDEIRA'] = df1['BANDEIRA'].str.upper()
df1['REGIONAL'] = df1['REGIONAL'].str.upper()
df1['MODELO'] = df1['MODELO'].str.upper()
df1['CATEGORIA'] = df1['CATEGORIA'].str.upper()
df1['SUBCATEGORIA'] = df1['SUBCATEGORIA'].str.upper()
df1['MARCA'] = df1['MARCA'].str.upper()
```
